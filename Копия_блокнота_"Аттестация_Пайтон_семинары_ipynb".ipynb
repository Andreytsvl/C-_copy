```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(60)
```

```python
data.loc[data['whoAmI'] == 'human'] = 1
data.loc[data['whoAmI'] == 'robot'] = 0
data['human'] = data['whoAmI'].apply(lambda x: x)
data['robot'] = data['whoAmI'].apply(lambda x: 1-x)
print(data)
```
